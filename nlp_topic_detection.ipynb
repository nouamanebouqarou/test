```python
from nltk.corpus import brown
import pandas as pd 

# Utilities for Exercice 1
def print_simples(text, title, maxlen = 3000):
    # used to compare and display two texts
    df = pd.DataFrame({title[0]:[text[0]], title[1]:[text[1]]})
    pd.set_option('display.max_colwidth', maxlen) 
    display(df)
    
#print(brown.categories())
corpusA = [] # 500 samples
for category in brown.categories():
    for fileid in brown.fileids(category):
        corpusA.append(' '.join(brown.words(fileids = fileid)))

print(len(corpusA))
#print(corpusA[0])
```

<div style = 'font-family:Garamond; font-size:120%;'><p>(2) Enlever les stop words, la ponctuation et les majuscules</p></div>

```python
from nltk.corpus import stopwords
stopWords = set(stopwords.words('English')) #get stopwords

import re
def remove_from_text(text, ignore_list, dont_keep = 'nps'):
    '''
    n : don't keep if it contains numbers
    p : don't keep if it contains punctuation
    s : don't keep if it contains whats in ignore_list
    '''
    if dont_keep == 'nps': regex = r"[0-9]+|[^\w\s]+|\b({})\b".format("|".join(ignore_list))
    elif dont_keep == 'ps': regex = r"[^\w\s]+|\b({})\b".format("|".join(ignore_list))
    elif dont_keep == 's': regex = r"\b({})\b".format("|".join(ignore_list))
    if dont_keep == 'np': regex = r"[0-9]+|[^\w\s]+"
    if dont_keep == 'n': regex = r"[0-9]+"
    if dont_keep == 'p': regex = r"[^\w\s]+"
    return re.compile(regex).sub(' ', text)
#test
#print(remove_from_text('?? hello 100 the he', stopWords, dont_keep='s'))

corpusA = [text.lower() for text in corpusA]# to lowercase
corpusB = [remove_from_text(text, stopWords, dont_keep='nps') for text in corpusA]# remove punctuation & stopwords
print_simples([corpusA[1], corpusB[1]], ['corpusA', 'corpusB'], maxlen = 1500)
```
